```python
import pandas as pd
raw=pd.read_csv("https://cocl.us/Geospatial_data")
raw
```

# Post code reindex by original post code 

```python
previous=pd.read_csv("a1.csv")
curr=previous.drop("Unnamed: 0",axis=1)
curr
```

```python
coord_df=pd.DataFrame([])
postcode=curr["Postcode"]

for i in postcode:
    coord_df=coord_df.append(raw[raw["Postal Code"]==i].iloc[:,1:])
coord_df=coord_df.reset_index(drop=True)

result=pd.concat([curr,coord_df],axis=1)
result.to_csv('toronto_loc.csv')
```

```python
import folium # map rendering library
map_newyork = folium.Map(location=[result["Latitude"][0], result["Longitude"][0]], zoom_start=10)
```

```python
map_newyork
```

```python
from sklearn.cluster import KMeans 
k_means = KMeans(init="k-means++", n_clusters=4, n_init=12)
k_means.fit(result[['Latitude','Longitude']])
k_means_labels = k_means.labels_
k_means_labels


k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
# add markers to map
fill_colors=['#3186cc','#824031','#FFE700','#1ED700']
c=0
for lat, lng, borough, neighborhood in zip(result['Latitude'], result['Longitude'], result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color=fill_colors[k_means_labels[c]],
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    c=c+1
map_newyork
```
